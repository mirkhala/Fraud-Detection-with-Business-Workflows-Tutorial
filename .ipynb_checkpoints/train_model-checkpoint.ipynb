```python
s3 = boto3.client(service_name='s3',
                  aws_access_key_id = s3_access_key,
                  aws_secret_access_key = s3_secret_key, 
                  endpoint_url=s3_endpoint_url)

key = "uploaded/creditcard-sample10k.csv"
obj = s3.get_object(Bucket=s3_bucket, Key=key)
df = pd.read_csv(io.BytesIO(obj['Body'].read()))
df.head()
```

```python
df.describe().T
```

```python
df.sort_values(by='Time', inplace=True)
df.head()
```

```python
df_train, df_test = train_test_split(df, test_size=0.25, random_state=42)
```

```python
df_train.head()
```

```python
columns_to_drop = [
    'Unnamed: 0',
    'Time', 
    'Class',
    'V1',
    'V2',
    'V5',
    'V6',
    'V7',
    'V8',
    'V9',
    'V13',
    'V15',
    'V16',
    'V18',
    'V19',
    'V20',
    'V21',
    'V22',
    'V23',
    'V24',
    'V25',
    'V26',
    'V27',
    'V28'
]

features_train = df_train.drop(columns_to_drop, axis=1)
target_train = df_train['Class']

features_test = df_test.drop(columns_to_drop, axis=1)
target_test = df_test['Class']
```

```python
features_train.head()
```

```python
features_test.head()
```

```python
model = RandomForestClassifier(n_estimators=200, 
                               max_depth=6, 
                               n_jobs=10,
                               class_weight='balanced')

model.fit(features_train, target_train)
print("Score:", model.score(features_test, target_test))
```

```python
#save mode in filesystem
dump(model, 'model.joblib') 
```

```python
key = "uploaded/model/model.joblib"
s3.upload_file(Bucket=s3_bucket, Key=key, Filename="model.joblib")
```
